```python
# @title Japanese Test - CT Analysis
jpn_cta_prompt_start = """Generate a centering theory analysis based on the provided Japanese text.

When writing a centering theory analysis, you must keep track of four features in each sentence:
    1. Cb (Backward-looking Center)
       The discourse entity that the current sentence is most centrally about, carried over from the previous sentence.
       It represents continuity with prior discourse.

    2. Cf (Forward-looking Centers)
       A ranked list of entities mentioned in the current sentence, ordered by their prominence (usually subject > object > others).
       These are potential candidates to become the center in the next sentence.

    3. Cp (Preferred Center)
       The highest-ranked element on the Cf list.
       It is the entity most likely to be continued as the focus of the next sentence.

    4. Transition
       Describes how the discourse focus shifts from one sentence to the next, based on Cb and Cp:
            1. Continue: Cb stays the same and equals Cp
            2. Retain: Cb stays the same but Cp differs
            3. Shift: Cb changes

Here is an example of a Japanese story and its corresponding centering theory analysis:
白い斜面を滑り降りながら、冷たい空気が肺の奥までしみ込むのを感じていた。
少し先では赤いジャケットを着た友人が、勢いよくターンを決めていた。
バランスを取りながら、コース脇に立っている青い旗をちらりと見た。
雪の中で鮮やかに際立っていた。
友人はスピードを落とすと、旗の近くで軽くジャンプして笑った。
こんな瞬間がずっと続けばいいのにと、心の奥で静かに願っていた。

Sentence 1:
    Cb: N/A
    Cf: [Narrator]
    Cp: Narrator
    Transition: N/A
Sentence 2:
    Cb: Narrator
    Cf: [Friend, Narrator, Jacket, Turn]
    Cp: Friend
    Transition: Shift
Sentence 3:
    Cb: Friend
    Cf: [Narrator, Friend, Flag]
    Cp: Narrator
    Transition: Retain
Sentence 4:
    Cb: Narrator
    Cf: [Flag, Snow]
    Cp: Flag
    Transition: Shift
Sentence 5:
    Cb: Flag
    Cf: [Friend, Flag, Jump]
    Cp: Friend
    Transition: Shift
Sentence 6:
    Cb: Friend
    Cf: [Narrator, Moment]
    Cp: Narrator
    Transition: Shift

Here is another example:
コンサートホールの静けさに耳を澄ませながら、胸の鼓動を感じていた。
翔太はゆっくりとピアノの前に座り、鍵盤に指を置いた。
柔らかな音を響かせて、こちらにちらりと視線を送った。
息を呑み、体が少し前のめりになった。
美咲は楽譜をめくり、ページを丁寧に押さえた。
その指先が紙をなぞりながら、リズムに合わせて軽く揺れた。
私は翔太に「その曲、ずっと聴きたかったんだ」とささやいた。
翔太は「じゃあ、最後まで楽しんで」と微笑んだ。
そのまま音を重ね、ホールいっぱいに旋律を広げていった。

Sentence 1:
    Cb: N/A
    Cf: [Narrator]
    Cp: Narrator
    Transition: N/A
Sentence 2:
    Cb: Narrator
    Cf: [Shota, Narrator, Piano]
    Cp: Shota
    Transition: Shift
Sentence 3:
    Cb: Shota
    Cf: [Shota, Narrator, Sound]
    Cp: Shota
    Transition: Continue
Sentence 4:
    Cb: Shota
    Cf: [Narrator, Shota]
    Cp: Narrator
    Transition: Retain
Sentence 5:
    Cb: Narrator
    Cf: [Misaki, Narrator, Sheet Music]
    Cp: Misaki
    Transition: Shift
Sentence 6:
    Cb: Misaki
    Cf: [Misaki, Paper, Rhythm]
    Cp: Misaki
    Transition: Continue
Sentence 7:
    Cb: Misaki
    Cf: [Narrator, Shota, Song]
    Cp: Narrator
    Transition: Shift
Sentence 8:
    Cb: Narrator
    Cf: [Shota, Narrator]
    Cp: Shota
    Transition: Shift
Sentence 9:
    Cb: Shota
    Cf: [Shota, Melody, Hall]
    Cp: Shota
    Transition: Continue

Now, you will write a centering theory analysis like the examples above based on the following text:\n"""

jpn_cta_prompt_end = "\n\nPlease respond only with your centering theory analysis."
```
